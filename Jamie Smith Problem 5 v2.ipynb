## 5. Valentine's Gift

Suppose you run a business creating custom picture frames, and Valentine's Day is fast approaching - the busiest day of the year for you.

To help handle the extra load, you want to create a program that automates the order placing process using Python.

You offer two different types of custom frames - Platinum and Plastic. 

For Platinum frames, you offer engraving, and a customer can enter _any text_ they want to be engraved on the frame.

For Plastic frames, they come in two sizes: Pocket and Standard.

Your goal is to write a program that gets a user's order - what type of frame they want (platinum or plastic), what the engraving should be (if it's platinum ONLY), and what size it should be (if it's plastic ONLY). Then, summarize the user's order back to them at the end of the program.

Here is what a successful solution could look like:

```
Welcome!

If you would like a Platinum frame, please type 1. If you would like a plastic frame, please type 2.

1

Would you like your frame engraved? (Please enter Yes or No, case sensitive):

Yes

What would you like the engraving to read?

Happy Valentine's Day!

Your Order:

Platinum Frame, With Engraving that reads Happy Valentine's Day
```

**TIP: For taking in options such as frame size or frame material, try and use a menu system so you don't need to worry about comparing text. For example, you could have it display something like this:**

````
print("Welcome! \n Please enter one of the following options: \n 1. Platinum Frame \n 2. Plastic Frame")
#Platinum or Plastic
FrameType = input("Frame Type: ")
if (FrameType != "1") and (FrameType != "2"):
  print("INVALID")
#If Plastic
elif (FrameType == "2"):
  print("Please enter one of the following options: \n 1. Pocket Size \n 2. Standard Size")
  FrameSize = input("Plastic Frame Size: ")
  if (FrameSize != "1") and (FrameType != "2"):
    print("INVALID")
  elif FrameSize == "1":
    print("Your Order: \n Plastic Frame, Pocket Size")
  else:
    print("Your Order: \n Plastic Frame, Standard Size")
#If Platinum
else:
  FrameMessage = input("Would you like your frame engraved? (Please enter Yes or No, case sensitive): ")
  if (FrameMessage != "Yes") and (FrameMessage != "No"):
    print("INVALID")
  elif FrameMessage == "No":
    print("Your Order: \n Platinum Frame, Without Engraving")
  else:
    Engraving = input("What would you like the engraving to read?")
    print("Your Order: \n Platinum Frame, With Engraving that reads",Engraving)


````
