## Warm-Up #1: Key Things to Remember About Lists

- Syntax uses **square brackets** `[ ]`

`myList = [400, 500, 150]`

- *Lists* are **ordered**

`[3, 4, 5] is NOT equal to [4, 5, 3]`

- **Number of elements** in a *list* called `myList`?

`len(myList)`

- Like *strings* and *tuples* that we saw earlier, individual *list* items can be accessed via **integer indexes** beginning with **index 0**.

`myList = [ "Ravens", "Eagles", "Washington Football", "Cowboys"]`

`print(myList[1]) # Prints Eagles`

- *Lists* can have arbitrary items

`arbList = [5, "Hello", False, (1,2,3), ["Ravens", "Eagles"] ]`

- *Lists* are **dynamic** - items can be added, deleted, and modified.

`movies = [ "42", "Lincoln", "The Natural", "Monsters, Inc." ]`

`print(movies[1:3])`

`movies.append("The Wizard of Oz")`

`del movies[2]`

`movies[3] = "Despicable Me"`

#### Try the code below:


````
movies = [ "42", "Lincoln", "The Natural", "Monsters, Inc." ]
print("The movies list: ", movies)

print("The second and third elements: ", movies[1:3])

movies.append("The Wizard of Oz")
print("The movies list after adding an item: ", movies)

del movies[2]
print("The movies list after deleting an item: ", movies)

movies[3] = "Despicable Me"
print("The movies list after replacing an item: ", movies)
````

## 3. Checking the Cart (pt. 1)

Suppose you're a new developer at a company who wants to create a program that lets the user search their shopping cart for multiple items.

You're taking over for the previous developer, who left you with some code to start with, shown below.

For the first part of this problem, we're going to do the following:

* Create a second **list** called `searchterms` to store **every** "search term" in.
* Create a **loop** that will keep asking the user to enter an item to search for _until_ the user types "STOP"
* On each iteration of the loop, store the item in the list.
* When the loop ends, print out the list

Here's what a successful solution should look like:

```
Enter an item to search for: avocados
Enter an item to search for: potatoes
Enter an item to search for: STOP
['avocados', 'potatoes']
```

**TIP: You might find it tricky to create the loop. You do not want to add an "STOP" to the search list. Consider using an `if` statement inside the loop to determine if it's appropriate to store the value or not.**

````
# Setup the Cart
my_cart = ["carrots", "avocados", "tomatoes", "potatoes", "green beans", "corn", "squash", "lima beans", "apples", "grapes"]

#searchterms
searchterms = []

searchFor = "go"

#loop
while searchFor != "stop":
  # Take in user input
  searchFor = input("Enter an item to search for: ")
  searchFor = searchFor.lower()
  if searchFor in my_cart:
    searchterms.append(searchFor)

#final
print(searchterms)

````

````
# Setup the Cart
my_cart = ["carrots", "avocados", "tomatoes", "potatoes", "green beans", "corn", "squash", "lima beans", "apples", "grapes"]

#searchterms
searchterms = []

searchFor = "go"

#loop
while searchFor != "stop":
  # Take in user input
  searchFor = input("Enter an item to search for: ")
  searchFor = searchFor.lower()
  searchterms.append(searchFor)

#remove stop
searchterms.remove("stop")

#print result
for term in searchterms:
  if term in my_cart:
    print(f"Found {term} in the cart")
  else:
    print(f"Did not find {term} in the cart")
    

````

## 5. Being Unique

Imagine that you collect state quarters - each state has a unique pattern on it's quarter, and you'd like to show which states you've collected so far.

However, as is the nature of finding quarters, you're bound to have more than one from certain states. Living in Maryland means that Maryland quarters should be abundant to you.

However, when you're displaying _which_ state you own a quarter from, you don't want to repeat the same state more than once.

So it's important to be able to remove duplicates from a list like this when printing.

Here's what your program should do:
* Create a list called `alreadyDisplayed`
* Walk through each state's name in `myQuarters` in a loop
* Check to see if that state's name is in `alreadyDisplayed`
* If it is in `alreadyDisplayed`, don't display anything
* If it is **NOT** in `alreadyDisplayed`, display the state's name to the screen and add it to `alreadyDisplayed`.

Here's what a successful solution should output:
```
myQuarters = ["Maryland", "Pennsylvania", "Maryland", "New York", "California", "Texas", "California", "Maryland"]

Maryland
Pennsylvania
New York
California
Texas
```

````
myQuarters = ["Maryland", "Pennsylvania", "Maryland", "New York", "California", "Texas", "California", "Maryland"]

alreadyDisplayed = []

for quarter in myQuarters:
  if quarter not in alreadyDisplayed:
    alreadyDisplayed.append(quarter)
    print(quarter)
````

## 6. High Score (CHALLENGE)

Suppose you're working for a company that programs arcade cabinets. They ask you to create a display that shows the top five highest scores from a list of scores.

Your solution should do the following:

* Re-order `allScores` so that they are stored lowest-to-highest.
* Go through `allScores` and print them to the screen, from **highest-to-lowest**.

**TIP #1: We know that calling `sort()` will return the scores from lowest-to-highest. Where will the top 5 scores be in your list?**

**TIP #2: Remember that `range()` can take in stride, just like slicing notation. This means we can move backwards through a list in reverse order. If you can select the range you want, all you need to do is use stride to walk through that range backwards.**

**TIP #3: If you're unsure how to print the top five scores from highest-to-lowest, try and print them from lowest-to-highest first.**

**TIP #4: How do you determine the length of `allScores`? How would you determine the index of the last five scores?**

````
allScores = [90,101,16,29,92,17,18,101,78,203,404,302,230,389,301,291]
allScores.sort()

#get last five
final = (len(allScores) - 6)

#print backwards
for score in allScores[::-1]:
  if allScores.index(score) > final:
    print(score)
````
