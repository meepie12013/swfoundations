### **Exercise - JSON**
- Use your editor or Google Docs to compose a JSON Object that has the following attributes:
  - `firstName`
  - `lastName`
  - `age`
  - `address` --- This should be its own object with **3 member**: `streetAddress`, `street`, `zipcode`
  - `ownsPet` (either true or false)
  - An array of `hobbies`
- Use [JSON Lint](https://jsonlint.com/) to check if it's valid JSON.

````
{
	"profile": {
		"firstName": "Jamie",
		"lastName": "Smith",
		"age": 25,
		"address": {
			"streetAddress": "1000",
			"street": "Hilltop Road",
			"zipcode": 21250
		},
		"ownsPet": true,
		"hobbies": ["design", "sleeping", "piano"]
	}
}
````
