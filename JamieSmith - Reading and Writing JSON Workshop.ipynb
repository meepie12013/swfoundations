#### Read in CSV Data.
- Be sure to import both the `csv` and `json` modules.
- Create 2 string variables with the names of the two files above.
- Create a list variable called `parklist` that will store all the NPS park data that you read into Python.

#### Save Python data as a JSON file.
- Create a dictionary called `nationalparks`.
- Create and populate two keys in this dictionary called `parks` and `numberofparks`.
- Use `json.dump()` to write this dictionary to `nationalparks.json`.

#### After successfully creating your JSON, download it and open it in Firefox to check your final data.



```
import csv
import json

#Create 2 string variables with the names of the two files above.
csvFile = "national_park_service.csv"
jsonFile = "national_park_service.json"

#------------------------------------------------------------------------

#Create a list variable called parklist that will store all the NPS park data that you read into Python.
parklist = []

#Open the CSV file for reading
with open(csvFile, "r", errors='ignore') as csv_file:
  readcsv = csv.DictReader(csv_file)
  for row in readcsv:
      parklist.append(row)

#Create a dictionary called nationalparks.
nationalparks = {}

#Create and populate two keys in this dictionary called parks and numberofparks.
nationalparks['parks'] = parklist
nationalparks['numberofparks'] = len(parklist)

#Use json.dump() to write this dictionary to nationalparks.json.
with open("nationalparks.json", "w") as newdata:
    json.dump(nationalparks, newdata, indent=3, sort_keys=True)
```
