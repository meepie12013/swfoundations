## 1. Keeping Count

In the code cell below is a program that runs through the numbers 1 to 100 and determines if each number is divisible by 2, divisible by 3, and divisible by 5.

Your job is to create a dictionary called `divisibleCount` that has three keys - 2, 3, and 5. The keys should be **integers**

The value for each of these key-value pairs should be zero. This is because the value will represent how many numbers between 1 and 100 are divisible by the key.

After creating the dictionary, modify the program below to count the **number of iterations in which the loop variable is divisible by each number in the dictionary**.

Finally, uncomment the three lines of code at the bottom of the code cell that print the number of each of these counts.

A successful solution should display the following:
```
Numbers divisible by 2: 50
Numbers divisible by 3: 33
Numbers divisible by 5: 20
```

**TIP: You will NOT need to use elif or else in this assignment**

````
# Create the dictionary
divisibleCount = {
    2: 0,
    3: 0,
    5: 0
}

# Go through 1 - 100
# Determine it's divisibility
for number in range(1, 101):
  if (number % 2) == 0:
    divisibleCount[2] += 1
  if (number % 3) == 0:
    divisibleCount[3] += 1
  if (number % 5) == 0:
    divisibleCount[5] += 1

# Keep track of successful divisibility in the dictionary

print(f"Numbers divisible by 2: {divisibleCount[2]}")
print(f"Numbers divisible by 3: {divisibleCount[3]}")
print(f"Numbers divisible by 5: {divisibleCount[5]}")
````

## 6.  Finding the largest value in a Dictionary [CHALLENGE]

There are quite a few ways to do this.  But write some code that displays the team with the highest win total in 2019.

````
winTotals = {'Phillies': 81, 
             'Rays': 96, 
             'Rangers':78, 
             'Athletics': 97, 
             'Nationals': 93,
             'Yankees':103,
             'Dodgers':106
             }

reverseWins = {}
Numbers = []

#reversing and adding to something that can be sorted
for Name, Score in winTotals.items():
  reverseWins[Score] = Name
  Numbers.append(Score)

#reversing scores
Numbers.sort()
Numbers.reverse()

#print the highest
print(reverseWins[Numbers[0]])
````
