```
import csv

#Read the entire CSV file using DictReader.
#read and add the entire dataset into a Python list named netflix_shows

netflix_shows = []

#Display all the contents first
#read each line into a list
with open('workshops_csvdata_netflix_titles.csv') as csv_file:
    readcsv = csv.DictReader(csv_file)
    for row in readcsv:
        netflix_shows.append(row)

#counter
counter = 0

#Display only the first 3 rows of the datafile.
while counter < 3:
  print(f'Row {counter + 1}: {netflix_shows[counter]}')
  counter += 1

#------------------------------------------------------------------------

#How many shows are in the dataset?
print(f"There are {len(netflix_shows)} shows listed")

#Display the 100th dictionary in the list.
getHundredth = 99
print(f'Show {getHundredth + 1}: {netflix_shows[getHundredth]}')

#------------------------------------------------------------------------

#Display the total number of "Movie" records in the dataset.
movies = 0
tvShows = 0
for n in netflix_shows:
  if n['type'] == "Movie":
    movies += 1
  elif n['type'] == "TV Show":
    tvShows += 1

#Be sure to use lower() to normalize the data ?? Why??
print(f"{movies} of the {len(netflix_shows)} netflix items are movies.")

#Display the number of "TV Show" records there are in the dataset.
print(f"{tvShows} of the {len(netflix_shows)} netflix items are TV shows.")

#------------------------------------------------------------------------

#How many Movies was Jonah Hill in? Which ones?
JonahMovies = []

#You should use str.find("Jonah Hill") to determine if he appears in the "cast" field. What does find() return if the string is NOT found? Of course, you'll need a counter too.
#       ((((     I used the "in" function     ))))
#After you've counted the number of Movies that he is in, then use append() to collect the movie titles that he is in.
#       ((((     I counted after appending rows     ))))
#Now do something similar and collect the entire ROW of data that he appears in.

for n in netflix_shows:
  if "Jonah Hill" in n['cast']:
    JonahMovies.append(n)

print(f"Jonah Hill is in {len(JonahMovies)} Movies, including:", end = ' ')
for n in JonahMovies:
  print(n['title'], end = ', ')

#Create a CSV called jonah_hill_netflix.csv and save all the rows of the movies that he is in.

with open('jonah_hill_netflix.csv', "w") as csv_out:
    # Defining the order of and column names of data in output CSV
    fields = ["show_id", "type", "title", "director", "cast", "country", "date_added", "release_year", "rating", "duration", "listed_in", "description"]
    # csv_writer = csv.DictWriter(csv_out, fieldnames=fields, dialect='unix')
    csv_writer = csv.DictWriter(csv_out, fieldnames=fields)

    csv_writer.writeheader()
    for movie in JonahMovies:
        csv_writer.writerow(movie)

#to check if it worked
#with open('jonah_hill_netflix.csv') as newFile:
#    readcsv = csv.DictReader(newFile)
#    for row in readcsv:
#        print(row)

#------------------------------------------------------------------------

#*[CHALLENGE]**
#How many Movies or TV shows was Jonah Hill in? Which ones?

JHmovies = 0
JHtvShows = 0
for n in JonahMovies:
  if n['type'] == "Movie":
    JHmovies += 1
  elif n['type'] == "TV Show":
    JHtvShows += 1

#Be sure to use lower() to normalize the data ?? Why??
print()
print(f"{JHmovies} of the {len(JonahMovies)} Jonah Hill items are movies.")

#Display the number of "TV Show" records there are in the dataset.
print(f"{JHtvShows} of the {len(JonahMovies)} Jonah Hill items are TV shows.")

#------------------------------------------------------------------------

#Make your program interactive and ask the user to input an actress/actor of their choice.

#Set up custom thingy
ActorMovies = []

Actor = input("Enter the first and last name of an actor: ")
Actor = Actor.title()

for n in netflix_shows:
  if Actor in n['cast']:
    ActorMovies.append(n)

print(f"{Actor} is in {len(ActorMovies)} Movies, including:", end = ' ')
for n in ActorMovies:
  print(n['title'], end = ', ')

#------------------------------------------------------------------------
```

```

#What's the longest duration movie in the dataset? The shortest? Display the titles.

movieTimes = []

#take durations
for n in netflix_shows:
  if n['type'] == "Movie":
    durationNum = n['duration']
    #get rid of "min"
    durationNum = durationNum[:-4]
    #make into an int so we can sort
    durationNum = int(durationNum)
    movieTimes.append(durationNum)

#sort them hoes
movieTimes.sort(reverse=True)

#turn it back into a string
winner = (str(movieTimes[0]) + " min")
print(winner)

#pull the movie of that length
for n in netflix_shows:
  if n['duration'] == (winner):
    print(f"At {winner}, {n['title']} is the longest movie!")
```
