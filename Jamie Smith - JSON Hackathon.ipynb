```
import json

#Read the file into your program using json.load().

oneMovie = "one_movie_data.json"
with open(oneMovie, "r") as jsondata:
    pydata = json.load(jsondata)

#------------------------------------------------------------------------

#What is the Python datatype of the data you just read in?

print(pydata)

#How many keys are in the top level object? What are they?
print(f"The keys are: ")
for key, value in pydata.items():
  print(key, end=", ")

#Capture the list of movies.

moviefile = "large_movies_full.json"

with open(moviefile, "r") as moviejson:
    movies = json.load(moviejson)

movie_list = movies["movies"]

#How many unique movies are captured by this JSON file?
print()
print()
print(f"There are {len(movie_list)} movies in the dictionary.")

#------------------------------------------------------------------------

#Count the number of "G-rated" movies by traversing the list of movies.
#What key are you looking at? What value are you matching against?
#Be sure to set a counter initialized with zero.

G_rated = 0

for n in movie_list:
  if n['Rated'] == "G":
    G_rated += 1

print()
print(f"There are {G_rated} G-rated movies.")

#Create a list with ALL the "Ratings" and their count.

rating = 0
ratingList = {}

for n in movie_list:
  currentKey = n['Rated']
  currentKey = currentKey.upper()
  if currentKey not in ratingList.keys():
    ratingList[currentKey] = 1
  else:
    ratingList[currentKey] += 1

#Print out this list.

#------------------------------------------------------------------------

#Now create a dictionary with ALL the "Ratings"

print(ratingList)

#Now provide a count of the number of movies with each Rating.
#Now try to normalize data for case-sensitivity of "Ratings" by using lower()

for key, value in ratingList.items():
  print(f"There are {value} {key} movies.")

#Count the number of "PG" rated films.
print()
print(f"There are {ratingList['PG']} PG rated films.")

#Create a JSON data file with the "Titles" of all "PG" rated movies.
# Write a new JSON to disk

PG_Movies = []

for n in movie_list:
  if n['Rated'].upper() == "PG":
    PG_Movies.append(n['Title'])

num_pg = len(PG_Movies)

PG_json = {
    "pgmovies":PG_Movies,
    "num_pg_movies": num_pg
}

#Your final JSON data file called pg_movie_titles.json should contain 2 keys: "pgmovies" and "num_pg_movies".

with open("pg_movie_titles.json", "w") as newdata:
    json.dump(PG_json, newdata)

#------------------------------------------------------------------------

cleanMovies = []

#Clean the data from large_movies_full.json by removing all movies with an "X" rating.
for n in movie_list:
  if n['Rated'].upper() != "X":
    cleanMovies.append(n)  

#Be sure the final, cleaned JSON has the following 3 keys: "movies", "filename", and "num_movies".
cleaned_json = {
    "movies":cleanMovies,
    "filename":"large_movies_cleaned.json",
    "num_movies":len(cleanMovies)
}

#Save the resulting data as JSON in large_movies_cleaned.json
with open("large_movies_cleaned.json", "w") as newdata:
    json.dump(cleaned_json, newdata)


```
